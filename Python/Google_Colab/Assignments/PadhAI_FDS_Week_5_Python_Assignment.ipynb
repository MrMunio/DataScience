````
#exercise 2: (problem 1) identifying differentiation slope
import math
def calc_func_slope(func,x_a,d_x=0.000001):
  f_x_a=func(x_a)
  f_x_a_delta=func(x_a + d_x)
  slope = (f_x_a_delta-f_x_a)/d_x
  return (slope,x_a,f_x_a,d_x)

def sqfunc(x):
  return x**2

print(calc_func_slope(sqfunc,5))
````

````
#exercise 2: (problem 3)

def generate_student_data(n):
  '''
  gets student data for n no of students and return the data in list
  of dictionary format
  '''
  students_data=[]
  for i in range(n):#gets every students details
    student_dict={}
    student_dict["Name"]=str(input("Enter {}st student's name :".format(i+1)))
    student_dict["Marks"]=[]
    for j in range(10):#gets each students marks in 10 subjects
      student_dict["Marks"].append(int(input("Enter {}'s marks in subject {} :".format(student_dict["Name"],j+1))))
    students_data.append(student_dict)
  return students_data

import random
def generate_random_student_data(n):
  '''
  generates randomly the student data without the unput from the user
  '''
  students_data=[]
  for i in range(n):
    student = {
        "Name" : "student{}".format(i+1),
        "Marks" : [random.randint(50,100) for _ in range(10)]
    }
    students_data.append(student)
  return students_data

def marks_operation(student,beta=0.99):
  '''
  takes dictionary object of a particular student as argument and
  returns a list of calculated marks according to prob statement
  '''
  marks_out=[1,]
  for i in range(10):
    temp=beta*marks_out[i]+(1-beta)*student["Marks"][i]
    marks_out.append(temp)
  return marks_out

sd=generate_random_student_data(3)
mo=marks_operation(sd[0])
print(sd)
print(mo)
````

````
#exercise 2: (problem 4)

def is_prime(n):
  if n<=1 :
    return False
  is_prime=True
  for i in range(2,n):
    if n%i==0 :
      is_prime=False
  return is_prime

def generate_fibonacci_set(n):
  '''
  generates a set of fibonacci sequence up to number 'n'
  '''
  my_list=[1,1,]
  index=2
  while my_list[index-1]<=n:
    my_list.append(my_list[index-1]+my_list[index-2])
    index +=1
  return set(my_list)

def create_number_dict(n):
  '''
  creates a dictionary with key as numbers from 1 to n. each key of number
  correspond to a list of strings stating that nuber's property(fibonacci,prime,square).
  '''
  number_dict={}
  fibonacci=generate_fibonacci_set(n)
  for i in range (1,n+1):
    if i not in number_dict:
      number_dict[i]=[]
    if is_prime(i):
      number_dict[i].append("Prime")
    if i in fibonacci:
      number_dict[i].append("Fibonacci")
    if i**2<=n:
      if i!=1:
        number_dict[i**2]=[]
      number_dict[i**2].append("Square")
  return number_dict

def search_number_dict(number_dict,property):
  '''
  takes number dictionary and any 1 property among prime, fibonacci, square
  and returns a touple of numbers having that property.
  '''
  number_list=[]
  for i in range(len(number_dict)):
    if property in number_dict[i+1]:
      number_list.append(i+1)
  return number_list

number_dict=create_number_dict(100)
print(search_number_dict(number_dict,"Fibonacci"))
print(search_number_dict(number_dict,"Square"))
print(search_number_dict(number_dict,"Prime"))
````

````
#exercise 2: (problem 5)

def encode_char(char):
  return chr(2*ord(char)+3)

def decode_char(e_char):
  return chr(int((ord(e_char)-3)/2))

text="Data science is an inter-disciplinary field that uses scientific methods, processes, algorithms and systems to extract knowledge\
 and insights from structured and unstructured data. Data science is related to data mining and big data. Data science is a \"concept to\
  unify statistics, data analysis, machine learning and their related methods\" in order to \"understand and analyze actual phenomena\" \
  with data. It employs techniques and theories drawn from many fields within the context of mathematics, statistics, computer science, \
  and information science. Turing award winner Jim Gray imagined data science as a \"fourth paradigm\" of science (empirical, theoretical,\
  computational and now data-driven) and asserted that \"everything about science is changing because of the impact of information \
  technology\" and the data deluge. In 2015, the American Statistical Association identified database management, statistics and machine \
  learning, and distributed and parallel systems as the three emerging foundational professional communities."

e_text=""
for char in text:
  e_text +=encode_char(char)
with open('encoded.txt','w') as f:
  f.write(e_text)

with open('encoded.txt','r') as f:
  encoded=f.read()
print("encoded file contents : ",encoded)

d_text=""
for char in encoded:
  d_text +=decode_char(char)

with open('decoded.txt','w') as f:
  f.write(d_text)

with open('decoded.txt','r') as f:
  decoded=f.read()
print("decoded file contents : ",decoded)
````
