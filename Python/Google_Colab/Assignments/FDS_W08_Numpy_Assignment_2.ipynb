```
import time
start_time = time.time()

out=np.zeros((row1,col2))
for r in range(row1):#3   3,2  2,5
  for c in range(col2):#5
    sum=0
    for x in range(col1):#2
      sum += a[r,x]*b[x,c]
    out[r,c]=sum

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Processing time: {elapsed_time:.4f} seconds")
out
```
