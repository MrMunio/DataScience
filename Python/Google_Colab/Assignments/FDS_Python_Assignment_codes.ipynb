```python
#1. Write an iterative function to compute the factorial of a natural number

def factorial(number):
  product = 1
  for i in range(number):
    product = product * (i+1)
  return product

print(factorial(6))
```

```python
#2. Write an recursive function to compute the factorial of a natural number
def factorial_rec(number):
  if number ==1:
    return 1
  else :
    return number * factorial_rec(number - 1)

print(factorial(5))
```
