#Exercise 1: finding no of points outside N_Dim unit sphere(i.e. area of unit square - area of one fourth circle embeded inside unit square )

```python
import numpy as np
```

```python
ndim=3
npoints=1000000
```

```python
#method 1:
%%time
points = np.random.rand(npoints,ndim)
dist_from_origin=np.zeros((npoints,1))
for i in range(npoints):
  for j in range(ndim):
    dist_from_origin[i] +=points[i,j]**2
count =0
for i in range(npoints):
  if dist_from_origin[i]>1 :
    count+=1
fraction_of_points_out_of_circle=count/npoints
print("are of area outside n dim unit sphere is: ",fraction_of_points_out_of_circle)
```

```python
#method 2:
%%time
points = np.random.rand(npoints,ndim)
points=np.square(points)
points=np.sum(points,axis = 1)# if axis is 1 then it does addition along the axis of column
fraction_of_points_out_of_circle=np.sum(points >1)/npoints
print("are of area outside n dim unit sphere is: ",fraction_of_points_out_of_circle)
```

```python
#method 2 single line notation:
%%time
points = np.random.rand(npoints,ndim)
fraction_of_points_out_of_circle=np.sum(np.sum(np.square(points),axis = 1)>1)/npoints
print("are of area outside n dim unit sphere is: ",fraction_of_points_out_of_circle)
```

```python
##method 2 single line function build:
def area_outside_sphere(npoints,ndim):
  points = np.random.rand(npoints,ndim)
  dist_from_origin=np.zeros((npoints,1))
  return np.sum(np.sum(np.square(points),axis = 1)>1)/npoints

for i in range(2,11):
  print(f'area out side sphere in {i} dimensions is : {area_outside_sphere(10**7,i)}')
```

```python
#common initialisations:
import numpy as np
cric_data=np.loadtxt("cric_data.tsv",skiprows=1,usecols=(1,2,3))
```

```python
#1.Find mean, median, IQR for Sachin, Rahul, and India
mean=np.mean(cric_data,axis=0)
median=np.median(cric_data,axis=0)
percentiles=np.percentile(cric_data,[25,75],axis=0)
mean=np.round(mean,2)
iqr=[]
for i in range (3):
  iqr.append([percentiles[1,i]-percentiles[0,i]])
print(f"mean median and iqr of Sachin is : {mean[0]}, {median[0]}, {iqr[0][0]} respectively")
print(f"mean median and iqr of Rahul is : {mean[1]}, {median[1]}, {iqr[1][0]} respectively")
print(f"mean median and iqr of India is : {mean[2]}, {median[2]}, {iqr[2][0]} respectively")
```

```python
#2.Find the histogram of Sachin's scores with 10 bins
sachin_data=cric_data[:,0]
np.histogram(sachin_data,bins=10)
```

```python
#3.Find mean of Sachin's scores for each groupe of 25 matches
np.mean(sachin_data.reshape(9,25),axis=1)
```

```python
#4.Find mean of Sachin's scores where he has scored a century
np.mean(sachin_data[sachin_data>=100])
```

```python
#5.Find mean of Sachin's scores when Rahul has scored less than 10
np.mean(cric_data[cric_data[:,1]<10][:,0]).round(2)
```

```python
#6.Find mean for Sachin's scores based on which quartile India's score falls in.(solution 1)
india_quartiles=np.percentile(cric_data[:,2],[25,50,75,100],axis=0)
india=cric_data[:,2]
for i in range(len(india_quartiles)):
  print(f"mean of sachin score in india's quartile {i+1} is\
  {np.mean(cric_data[india<=india_quartiles[i]][:,0]).round(2)} ")
```

```python
#6.(v2)Find mean for Sachin's scores based on which quartile India's score falls in.(solution 2)
india_quartiles=np.percentile(cric_data[:,2],[25,50,75,100],axis=0).reshape(4,1)
india=cric_data[:,2]
Indices=india<=india_quartiles
for i in range(4):
  print(f"mean of sachin score in india's quartile {i+1} is\
  {np.mean(sachin_data[Indices[i,:]]).round(2)} ")
```

```python
#7.For every match find out who has has scored more - Sachin or Rahul
is_sachin_higher=cric_data[:,0]>cric_data[:,1]
print(f"Match_No   High_Scorer")
for i in range(len(is_sachin_higher)):
  if is_sachin_higher[i]:
    print(f"{i+1}           Sachin")
  else:
    print(f"{i+1}           Rahul")
```

```python
#7.For every match find out who has has scored more - Sachin or Rahul
# (version 2: using "where" function)
is_sachin_higher=cric_data[:,0]>cric_data[:,1]
np.where(is_sachin_higher,"sachin","Rahul")
```

```python
#8.How many more runs does Sachin score on average after having scored x runs
x=20
np.mean(sachin_data[sachin_data>=x]).round(2)
```

```python
#9.How many matches did Sachin take to score first 1000 runs, next 1000 runs, ...
cum_sum_sachin=np.cumsum(sachin_data)
matches=0
multiple=1
for i in range(len(cum_sum_sachin)):
  matches += 1
  if cum_sum_sachin[i]>=1000*multiple:
    print(f"sachin scored his {1000*multiple} runs in his {matches}th match")
    multiple +=1
```

```python
#9.How many matches did Sachin take to score first 1000 runs, next 1000 runs, ...
#version 2: using histogram
cum_sum_sachin=np.cumsum(sachin_data)
np.cumsum(np.histogram(cum_sum_sachin,bins=np.arange(0,max(cum_sum_sachin)+1,1000))[0])+1
```
